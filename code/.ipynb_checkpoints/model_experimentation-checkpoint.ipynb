```python
import pandas as pd
from datetime import datetime
import numpy as np

from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.base import clone
import seaborn as sns
```

```python
train_df = pd.read_csv('../data/train_data.csv')
print(train_df.shape)
print(list(train_df.columns))
train_df.head()
```

```python
test_df = pd.read_csv('../data/test_data.csv')
print(test_df.shape)
print(list(test_df.columns))
test_df.head()
```

# Experiment 1

### Feature Engg
* few columns will be removed
* order_value, order_items_count will be kept for now and removed later. ***
* created weekday and hour features from timestamp and deleted timestamp
* created 'is_postcode_digit' from delivery postcode and deleted delivery_postal_code
* binary variables are kept as is for now. ***
* customer_id changed to 'num_times_ordered' and deleted customer_id
* delivery_geohash_precision8 changed to geohash_95quantile of dropoff distance in that geohash and deleted delivery_geohash_precision8.

```python
date_format = '%Y-%m-%dT%H:%M:%S'
date_to_weekday = lambda x:datetime.strptime(x, date_format).weekday()
date_to_hour = lambda x:datetime.strptime(x, date_format).hour

def common_feature_engineering(df):
    
    del df['order_id']
    del df['order_value'] # removing order_value and item_count for another experiment
    del df['order_items_count']
    
    df['is_postcode_digit']= df['delivery_postal_code'].apply(lambda x: str(x).isdigit()) #checking if postcode is a digit
    del df['delivery_postal_code']
    
    df['ts_weekday'] = df['created_timestamp_local'].apply(date_to_weekday).astype(np.int8) # converting to hour and weekday features
    df['ts_hour'] = df['created_timestamp_local'].apply(date_to_hour).astype(np.int8)
    del df['created_timestamp_local']
    return df
    
def post_split_feature_engg(df, customer_dict=None, geohash_dict=None, geohash_impute_val=None):
    # when train data set is given, customer_dict along with others values, will be None.
    # for test dataset, customer_dict, geohash_dict and geohash_impute_val will be the ones generated from the train dataset.
    
    if customer_dict is None: 
        df = df[df.logistics_dropoff_distance <= 20000]
        customer_dict = df['customer_id'].value_counts().to_dict()
        geohash_dict = df[['delivery_geohash_precision8', 'logistics_dropoff_distance']].groupby('delivery_geohash_precision8') \
                        .agg({'logistics_dropoff_distance': lambda x: x.quantile(0.95)}).to_dict()['logistics_dropoff_distance'] # taking 95 percentile of dropoff distance in a geohash code
        geohash_impute_val = df['logistics_dropoff_distance'].quantile(0.95) #geohash impute val is 95 percentile of all logistics dropoff distance
    
    df['num_times_ordered'] = df['customer_id'].apply(lambda x:customer_dict.get(x, 1)) # num times a customer ordered, default val = 1 
    df['geohash_95quantile'] = df['delivery_geohash_precision8'].apply(lambda x: geohash_dict.get(x, geohash_impute_val))
    
    del df['customer_id']
    del df['delivery_geohash_precision8']
    
    return df, customer_dict, geohash_dict, geohash_impute_val
```

```python
train_df = common_feature_engineering(train_df)
test_df = common_feature_engineering(test_df)
```

```python
train_df.head()
```

```python
# train_columns = ['has_instruction','has_phone_number',
#                 'has_call_instruction', 'has_leave_instruction', 'has_lift_instruction',
#                 'has_lobby_instruction', 'has_gate_instruction', 'has_knock_instruction',
#                 'has_bell_instruction', 'is_postcode_digit', 'ts_weekday', 'ts_hour',
#                 'num_times_ordered', 'geohash_95quantile']

train_columns = ['is_postcode_digit', 'ts_weekday', 'ts_hour',   # columns to use while training.
                'num_times_ordered', 'geohash_95quantile']


def get_processed_train_val_matrix(df, train_ind, val_ind):
        
    proc_train_df, customer_dict, geohash_dict, geohash_impute_val  = post_split_feature_engg(pd.DataFrame(df.iloc[train_ind,:], copy=True))
    proc_val_df, _, _, _ = post_split_feature_engg(pd.DataFrame(df.iloc[val_ind,:], copy=True), customer_dict, geohash_dict, geohash_impute_val)
    
    x_train = proc_train_df[train_columns].values
    y_train = proc_train_df['logistics_dropoff_distance'].values
    x_val = proc_val_df[train_columns].values
    y_val = proc_val_df['logistics_dropoff_distance'].values
    
    return x_train, y_train, x_val, y_val

def train_model_loop(model_str, train_df):
    
    tscv = TimeSeriesSplit(n_splits=3) # time series k fold
    format_row = "{:>15}" * 3
    print(format_row.format("", *['train rmse', 'val rmse']))
    
    for train_ind, val_ind in tscv.split(train_df):
        
        if model_str == 'xgbr':
            model = XGBRegressor(n_estimators=300, max_depth=3, n_jobs=-1)
        else:
            model = RandomForestRegressor(n_estimators=300, max_depth=3, criterion='mse' ,n_jobs=-1)
        
        x_train, y_train, x_val, y_val =  get_processed_train_val_matrix(train_df, train_ind, val_ind)

        model = model.fit(x_train, y_train)
        y_train_prediction = model.predict(x_train)
        y_val_prediction = model.predict(x_val)

        train_rmse = "{:.4f}".format(mean_squared_error(y_train, y_train_prediction, squared=False))
        val_rmse = "{:.4f}".format(mean_squared_error(y_val, y_val_prediction, squared=False))

        print(format_row.format('', *[train_rmse, val_rmse]))
    return model, x_train, y_train, x_val, y_val
```

```python
#training xgb
xgbr_data = train_model_loop('xgbr', train_df)
xgbr = xgbr_data[0]
```

```python
# training random forest
rfm_data = train_model_loop('rfm', train_df)
random_forest_model = rfm_data[0]
```

```python
ax = sns.barplot(x = train_columns, y=xgbr.feature_importances_)
ax.set(title='XGBoost Feature Importance')
_ = ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```

```python
ax = sns.barplot(x = train_columns, y=random_forest_model.feature_importances_)
ax.set(title='Random Forest Feature Importance')
_ = ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```

```python
# seeing how much better than mean value does our model perform
model, x_train, y_train, x_val, y_val = rfm_data
y_val_pred = model.predict(x_val)
mean_y_val = y_train.mean()
print(mean_squared_error(y_val, np.ones_like(y_val)*mean_y_val, squared=False))
```

```python
pd.Series(y_val_pred).describe()
```

```python
# relation of error with y_vals
df = pd.DataFrame(x_val, columns = train_columns)
df['y'] = y_val
df['residual'] = y_val - y_val_pred
df['y_bin'], bins = pd.qcut(df.y, 20, retbins=True)
df['y_bin'] = df['y_bin'].astype(str)
```

```python
sns.set(rc={"figure.figsize":(20, 10)}) 
ax = sns.violinplot(x='y_bin', y='residual' ,kind='violin' ,data=df )
ax.set(title='Error analyis')
_ = ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```

```python
#predicting for test data
train_dataframe, customer_dict, geohash_dict, geohash_impute_val = post_split_feature_engg(pd.DataFrame(train_df, copy=True))
test_dataframe, _, _, _ = post_split_feature_engg(pd.DataFrame(test_df, copy=True), customer_dict, geohash_dict, geohash_impute_val)
x_train = train_dataframe[train_columns].values
y_train = train_dataframe['logistics_dropoff_distance'].values
x_test = test_dataframe[train_columns].values
y_test = test_dataframe['logistics_dropoff_distance'].values

model = RandomForestRegressor(n_estimators=300, max_depth=3, criterion='mse' ,n_jobs=-1)
model = model.fit(x_train, y_train)

y_train_prediction = model.predict(x_train)
y_test_prediction = model.predict(x_test)

train_rmse = "{:.4f}".format(mean_squared_error(y_train, y_train_prediction, squared=False))
test_rmse = "{:.4f}".format(mean_squared_error(y_test, y_test_prediction, squared=False))

print("Train rmse =", train_rmse)
print("Test rmse =", test_rmse)
```
