```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
train_df = pd.read_csv('../data/train_data.csv')
print(train_df.shape)
print(list(train_df.columns))
train_df.head()
```

```python
#checking for null values
train_df.isnull().any()
```

```python
print('percent of nulls in postal code =',(train_df['delivery_postal_code'].isnull().sum()*100.0)/len(train_df))
```

```python
for col,type_col in zip(train_df.columns.values,train_df.dtypes.values):
    if type_col == 'bool':
        train_df[col] = train_df[col].astype(np.int8)
```

```python
# creating a unix timestamp column
date_format = '%Y-%m-%dT%H:%M:%S'
date_str_to_timestamp = lambda x: int(datetime.strptime(x,date_format).timestamp())
train_df['unix_timestamp'] = train_df['created_timestamp_local'].apply(date_str_to_timestamp)
```

```python
print(train_df.dtypes)
train_df.head()
```

```python
# histograms of some columns
columns = ['order_value', 'order_items_count', 'has_instruction', 
           'logistics_dropoff_distance', 'has_phone_number', 
           'has_call_instruction', 'has_leave_instruction', 'has_lift_instruction', 
           'has_lobby_instruction', 'has_gate_instruction', 'has_knock_instruction', 
           'has_bell_instruction', 'unix_timestamp']
fig, ax = plt.subplots(len(columns), figsize=(20,40))
fig.suptitle('Histograms of selected columns')
for i, col in enumerate(columns):
    ax[i].set_title(col)
    ax[i].hist(train_df[col], bins=50)    
```

```python
train_df[columns].describe().round(3)
```

### Findings
* Postal code has some null values
* Order value and order count follow the expected exponential distribution.
* Interestingly a large portion of orders have instructions available.
* All other bool variables are largely negative.
* logistics_dropoff_distance has some outliers due to which graph is not informational. Have to remove these outliers. The 75 percentile is just 6 meters which is not that much
* timestamp values have spikes. Might be due to day and night seasonality. There appears to be a another seasonality in the "background".

```python
# checking outliers in logistics_dropoff_distance
distance_m_list = [500,1000,2000,3000,10000]
for distance_m in distance_m_list:
    num_outliers = (train_df['logistics_dropoff_distance'] >= distance_m).sum()
    print('#Values greater than',distance_m,'meters =',num_outliers,': Percent =',(num_outliers*100.0)/len(train_df))
```

```python
# plotting histogram for logistics_dropoff_distance after removing some outliers
fig, ax = plt.subplots(1, figsize=(20,10))
fig.suptitle('Histograms of logistics_dropoff_distance after removing some outliers')
hist_y, hist_x,_ = ax.hist(train_df[train_df['logistics_dropoff_distance'] <= 2000]['logistics_dropoff_distance'],bins=500)
print("Mode of distribution at =",hist_x[np.argmax(hist_y)] )
```

### Findings
* logistics_dropoff_distance has mode at 19m, mean at 97m and median at 37m.
* some outliers are can be attributed to some fault in gps, courier newness etc. 
  But removing outliers might not be the right thing to do at this point. More analysis is needed.

```python
# timestamp analysis
print("Start and End of timeseries = ", min(train_df.created_timestamp_local),':',max(train_df.created_timestamp_local))
fig, ax = plt.subplots(1, figsize=(20,10))
fig.suptitle("Detailed timestamp graph")
_ = ax.hist(train_df.unix_timestamp, bins=500)
```

```python
date_to_weekday = lambda x:datetime.strptime(x, date_format).weekday()
date_to_hour = lambda x:datetime.strptime(x, date_format).hour
date_to_day = lambda x:datetime.strptime(x, date_format).day
fig, ax = plt.subplots(2,figsize=(20,20))
ax[0].set_title("Day of week histogram of orders")
ax[0].hist(train_df.created_timestamp_local.apply(date_to_weekday), bins=7)
ax[1].set_title("Hour of day histogram of orders")
_ = ax[1].hist(train_df.created_timestamp_local.apply(date_to_hour), bins=24)
```

### Findings
* Can see spike pattern for each day in the graph of the whole data.
* There is almost no pattern while plotting data on a weekly basis.
* On an hourly basis, there are two modes as expected.

```python
# creating pairplot of variables that might not be useful with dropoff_distance
columns = ['order_value','order_items_count', 'logistics_dropoff_distance']
sns.pairplot(train_df[train_df.logistics_dropoff_distance <= 4000][columns])
```

```python
#checking relation between delivery_postal_code and delivery_geohash_precision8
def agg_func(data):
    new_cols = {}
    new_cols['num_geohash'] = data['delivery_geohash_precision8'].unique().shape[0]
    return pd.Series(new_cols)

grouped_df = train_df[['delivery_postal_code','delivery_geohash_precision8']].groupby('delivery_postal_code').apply(agg_func)
print('Number of unique postal_code =',train_df['delivery_postal_code'].unique().shape[0])
print('Number of unique geohash codes =',train_df.delivery_geohash_precision8.unique().shape[0])
print('Sum of geohashes in the grouped_df =', grouped_df['num_geohash'].sum())
```

```python
#checking values of delivery_postal_code.
train_df['delivery_postal_code'].unique()[500:1000]
```

```python
#plotting distribution of logistics_dropoff_distance for each pincode and sorting by logistics_dropoff_distance
def distance_agg_func(data):
    new_cols = {}
    new_cols['mean'] = data['logistics_dropoff_distance'].mean()
    new_cols['count'] = len(data)
    new_cols['95percentile'] = data['logistics_dropoff_distance'].quantile(0.95)
    return pd.Series(new_cols)

grouped_df = train_df[train_df.logistics_dropoff_distance <= 20000][['delivery_postal_code','logistics_dropoff_distance']].groupby('delivery_postal_code').apply(distance_agg_func).sort_values('mean',ascending=False)

fig, ax = plt.subplots(3, figsize=(20,20))
fig.suptitle('plots of mean, count and 95 percentile')
ax[0].plot(np.arange(len(grouped_df)), grouped_df['mean'], 'b.')
ax[1].plot(np.arange(len(grouped_df)), grouped_df['count'], 'y.')
ax[2].plot(np.arange(len(grouped_df)), grouped_df['95percentile'], 'r.')
```

```python
#plotting distribution of logistics_dropoff_distance for each geohash and sorting by logistics_dropoff_distance
grouped_df = train_df[train_df.logistics_dropoff_distance <= 20000][['delivery_geohash_precision8','logistics_dropoff_distance']].groupby('delivery_geohash_precision8').apply(distance_agg_func).sort_values('mean',ascending=False)

fig, ax = plt.subplots(3, figsize=(20,20))
fig.suptitle('plots of mean, count and 95 percentile')
ax[0].plot(np.arange(len(grouped_df)), grouped_df['mean'], 'b-')
ax[1].plot(np.arange(len(grouped_df)), grouped_df['count'], 'y-')
ax[2].plot(np.arange(len(grouped_df)), grouped_df['95percentile'], 'r-')
```

```python
### checking stats for correct/incorrect pincodes.
train_df['is_postcode_digit'] = train_df['delivery_postal_code'].apply(lambda x: str(x).isdigit()).astype(np.int8)
df = pd.DataFrame()
df['zero'] = train_df[(train_df['is_postcode_digit'] == 0) & (train_df.logistics_dropoff_distance <= 30000)]['logistics_dropoff_distance'].describe().round(2)
df['one'] = train_df[(train_df['is_postcode_digit'] == 1) & (train_df.logistics_dropoff_distance <= 30000)]['logistics_dropoff_distance'].describe().round(2)
df
```

### Findings
* There are some geohash codes which fall under multiple pincodes, could be due to string and nan pincodes.
* There are pincodes which are not digits.
* There is a significant difference in dropoff distance when pincode is not a digit


```python
# customer analysis
print('Unique customers =', train_df['customer_id'].unique().shape[0])
train_df['order_counts'] = train_df.groupby('customer_id')['customer_id'].transform('count')
sns.catplot(x='order_counts', y='logistics_dropoff_distance', data=train_df[train_df.logistics_dropoff_distance <= 20000], height=10, aspect=2)
```

```python
# calculating mean of dropoff distance for a customer to reduce clutter of the graph above
def customer_agg_func(data):
    new_cols = {}
    new_cols['count'] = len(data)
    new_cols['mean_distance'] = data['logistics_dropoff_distance'].mean()
    return pd.Series(new_cols)

grouped_df = train_df[train_df.logistics_dropoff_distance <= 20000][['customer_id','logistics_dropoff_distance']].groupby('customer_id').apply(customer_agg_func).sort_values('mean_distance')
```

```python
sns.catplot(x='count', y='mean_distance', data=grouped_df,height=10, aspect=2)
```

### Findings
* Dropoff distance becomes less as a customer orders more.
* Less variance in mean_distance/drop_off distance for customers with more orders could be due to less number of such customers.
